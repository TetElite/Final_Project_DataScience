# Data Collection
This notebook handles loading the raw drug review data and filtering it for pain-related conditions and medications.

## 1. Import Required Libraries

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import os

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', 100)
```

## 2. Load Raw Data

```python
# TODO: Load the raw data from TSV file
# Path: ../../data/raw/drugsComTrain_raw.tsv
raw_data_path = '../../data/raw/drugsComTrain_raw.tsv'

# Load data using pd.read_csv with tab separator
df = pd.read_csv(raw_data_path, sep='\t')

print(f"Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(f"Total records: {len(df)}")
```

## 3. Explore Dataset Structure

```python
# Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

# Display dataset information
print("\nDataset Information:")
df.info()

# Display statistical summary
print("\nStatistical Summary:")
display(df.describe())

# Check column names
print("\nColumn Names:")
print(df.columns.tolist())
```

## 4. Define Pain Condition and Medication Filters

```python
# Define pain-related conditions
pain_conditions = [
    'headache', 'migraine', 'back pain', 'arthritis', 'sciatica', 
    'fibromyalgia', 'toothache', 'neck pain', 'joint pain', 
    'osteoarthritis', 'rheumatoid arthritis', 'chronic pain', 
    'neuropathic pain', 'muscle pain'
]

# Define pain medications
pain_medications = [
    'ibuprofen', 'acetaminophen', 'naproxen', 'aspirin', 'diclofenac', 
    'tramadol', 'hydrocodone', 'oxycodone', 'tylenol', 'advil', 
    'aleve', 'motrin', 'celebrex', 'meloxicam', 'indomethacin'
]

print(f"Pain conditions to filter: {len(pain_conditions)}")
print(f"Pain medications to filter: {len(pain_medications)}")
```

## 5. Filter Dataset

```python
# TODO: Filter dataset for pain-related conditions and medications
# Use case-insensitive regex pattern matching

# Create regex patterns
condition_pattern = '|'.join(pain_conditions)
medication_pattern = '|'.join(pain_medications)

# Filter by condition (case-insensitive)
condition_mask = df['condition'].str.contains(condition_pattern, case=False, na=False, regex=True)

# Filter by drug name (case-insensitive)
drug_mask = df['drugName'].str.contains(medication_pattern, case=False, na=False, regex=True)

# Combine filters (OR logic - either condition OR medication matches)
filtered_df = df[condition_mask | drug_mask].copy()

print(f"Records after filtering: {len(filtered_df)}")
print(f"Percentage retained: {(len(filtered_df)/len(df))*100:.2f}%")
```

## 6. Save Filtered Data

```python
# TODO: Save filtered data to CSV
output_path = '../../data/filtered/pain_meds_filtered.csv'

# Create directory if it doesn't exist
os.makedirs(os.path.dirname(output_path), exist_ok=True)

# Save to CSV
filtered_df.to_csv(output_path, index=False)

print(f"Filtered data saved to: {output_path}")
```

## 7. Summary Statistics

```python
# Display summary statistics
print("=" * 60)
print("DATA COLLECTION SUMMARY")
print("=" * 60)
print(f"Original dataset rows: {len(df):,}")
print(f"Filtered dataset rows: {len(filtered_df):,}")
print(f"Rows removed: {len(df) - len(filtered_df):,}")
print(f"Retention rate: {(len(filtered_df)/len(df))*100:.2f}%")
print("=" * 60)

# Display top conditions
print("\nTop 10 Pain Conditions:")
print(filtered_df['condition'].value_counts().head(10))

# Display top medications
print("\nTop 10 Pain Medications:")
print(filtered_df['drugName'].value_counts().head(10))
```
